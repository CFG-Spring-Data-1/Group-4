# altering the database 

```python
import pandas as pd 
```

```python
import zipfile
with zipfile.ZipFile('Fake.csv.zip', 'r') and zipfile.ZipFile('True.csv.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```python
df= pd.read_csv('Fake.csv')
df.head()
df_2 = pd.read_csv('True.csv')
df_2.head()
df_2['label']=1
df['label']=0
# zero is fake and one is real 
```

```python
df
```

```python
df_combine = df.append(df_2)
df_combine.head()
```

```python
import zipfile
with zipfile.ZipFile('FA-KES-Dataset.csv.zip', 'r') as zip_ref:
    zip_ref.extractall()
df_second_piece=pd.read_csv('FA-KES-Dataset.csv',encoding='unicode_escape')
df_second_piece.head()
```

```python
df_final_1 = df_combine[['title','text','label']].copy()
df_final_2 = df_second_piece[['article_title','article_content','labels']].copy()
df_final_1
```

```python
df_combine_final=pd.concat([df_final_1,df_final_2.rename(columns={'labels':'label', 'article_title':'title', 'article_content':'text'})], ignore_index=True)
```
